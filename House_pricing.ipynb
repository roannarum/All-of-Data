```python
import pandas as pd
data = pd.read_csv('/content/house_prices.csv')
data.head(10)
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['LotArea'],data['SalePrice'])
plt.title('NYC House Pricing')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```



*   Tren kenaikan harga lahan diiringi karena kenaikan luas lahan.
*   Ketika luas lahan semakin besar maka harga jual pul meningkat




Misalkan ingin prediksi harga lahan untuk lahan dengan luas 14000, bagaimana caranya?

```python
plt.scatter(data['LotArea'],data['SalePrice'])
plt.axvline(x=14000, linewidth='1',color='red')
plt.title('NYC HOUSE PRICING')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['LotArea'],data['SalePrice'])
plt.plot(data.LotArea, 30000+15*data.LotArea, color = 'red')
plt.plot(data.LotArea, 10000+17*data.LotArea, color = 'green')
plt.plot(data.LotArea, 50000+12*data.LotArea, color = 'yellow')
plt.title('NYC HOUSE PRICING')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# red line
print('red line:', 30000+15*14000) #mengetahui untuk harga tanah yang 14000 persamaan garis merah

# yellow line
print('yellow line:', 50000+12*14000) 
# green line
print('green line:', 10000+17*14000)
```

```python
# Regression Modelling
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
x = data.LotArea[:,np.newaxis]
y = data.SalePrice
```

```python
regressor = LinearRegression()
regressor.fit(x,y)

y_pred = regressor.predict(x)
```

```python
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color = 'red')
plt.title('NYC HOUSE PRICING')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color = 'red')
plt.title('NYC HOUSE PRICING')
plt.xlabel('Area')
plt.ylabel('Price')
plt.axvline(x=14000, linewidth='1',color='green')
plt.show()
```
